```python
#Crear CSV limpio desde los JSON (histórico, no se corre siempre)
ruta_crudos = Path("../datos/crudos")
archivos = sorted(ruta_crudos.glob("StreamingHistory_music_*.json"))
archivos
```

```python
#Crear CSV limpio desde los JSON (histórico, no se corre siempre)
lista_df = []
for archivo in archivos:
    print("Cargando:", archivo.name)
    df_temp = pd.read_json(archivo)
    lista_df.append(df_temp)

df = pd.concat(lista_df, ignore_index = True)
df.head()
```

```python
#Crear CSV limpio desde los JSON (histórico, no se corre siempre)
df = df.rename(columns={
    "endTime": "fecha_reproduccion",
    "artistName": "artista",
    "trackName": "cancion",
    "msPlayed": "ms_reproducidos"
})

df.head()
```

```python
#Crear CSV limpio desde los JSON (histórico, no se corre siempre)
df["fecha_reproduccion"] = pd.to_datetime(df["fecha_reproduccion"])
df["fecha_reproduccion"].head()
```

```python
#Crear CSV limpio desde los JSON (histórico, no se corre siempre)
df["minutos_reproducidos"] = df["ms_reproducidos"]/60000
f = df["fecha_reproduccion"]
df["anio"] = f.dt.year
df["mes"] = f.dt.month
df["dia"] = f.dt.day
df["dia_semana"] = f.dt.day_name()
df["hora"] = f.dt.hour

df[[
    "fecha_reproduccion", "artista", "cancion",
    "ms_reproducidos", "minutos_reproducidos",
    "anio", "mes", "dia", "dia_semana", "hora"
]].head()
```

```python
#Crear CSV limpio desde los JSON (histórico, no se corre siempre)
df["usuario"] = "Elias"
df[["usuario", "artista", "cancion"]].head()
```

```python
#Crear CSV limpio desde los JSON (histórico, no se corre siempre)
from pathlib import Path
ruta_salida = Path("../datos/procesados/elias_limpio.csv")
df.to_csv(ruta_salida, index=False)
ruta_salida
```

```python
#Crear CSV limpio desde los JSON (histórico, no se corre siempre)
df_limpio = pd.read_csv(ruta_salida)
df_limpio.head()
```

```python
#Versión exploratoria, antes de modular en consultas.py
top_artistas = (
    df.groupby("artista")["minutos_reproducidos"].sum().sort_values(ascending=False).head(20)
)
top_artistas
```

```python
#Versión exploratoria, antes de modular en consultas.py
horas_del_dia = (
    df.groupby("hora")["minutos_reproducidos"].sum().sort_values(ascending=False)
)
horas_del_dia
```

```python
#Versión exploratoria, antes de modular en consultas.py
por_dia_semana = (
    df.groupby("dia_semana")["minutos_reproducidos"].sum().sort_values(ascending=False)
)
por_dia_semana
```

```python
#Versión exploratoria, antes de modular en consultas.py
por_anio_mes = (
    df.groupby(["anio", "mes"])["minutos_reproducidos"]
      .sum()
      .reset_index()
      .sort_values(["anio", "mes"])
)

por_anio_mes.head(20)
```
